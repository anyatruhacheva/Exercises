```python
import numpy as np
import pandas as pd
df = pd.read_excel('Data/adult.data.xlsx') 
df.head()
```

```python
df['race'].value_counts()
```

```python
df.loc[df['sex'] == ' Male', 'age'].mean().round()
```

```python
(len(df.loc[df['education'] == ' Bachelors'])/ len(df)) * 100
```

```python
advanced_education = df.loc[((df['education'] == ' Bachelors') | (df['education'] == ' Masters')| (df['education'] == ' Doctorate')) & (df['salary'] == '>50K')].shape[0]
percentage_of_people_with_advanced_education_more_50k = (advanced_education / len(df)) * 100
print(percentage_of_people_with_advanced_education_more_50k)
```

```python
without_advanced_education = len(df) - df.loc[(df['salary'] == '<=50K')].shape[0] - df.loc[((df['education'] == ' Bachelors') | (df['education'] == ' Masters')| (df['education'] == ' Doctorate')) & (df['salary'] == '>50K')].shape[0]
percentage_of_people_without_advanced_education_more_50k = (without_advanced_education / len(df)) * 100
print(percentage_of_people_without_advanced_education_more_50k)
```

```python
df['hours-per-week'].min()
```

```python
number_of_hours = df['hours-per-week'].min()
percentage_of_people = ((df.loc[(df['hours-per-week'] == 1)& (df['salary'] == '>50K')].shape[0])/ df.loc[(df['hours-per-week'] == 1)].shape[0]) * 100
print(percentage_of_people)
```

```python
df.loc[df['salary'] == '>50K', 'native-country'].value_counts()
percentage = (df.loc[df['salary'] == '>50K', 'native-country'].value_counts() / df['native-country'].value_counts()) * 100
percentage.sort_values(ascending = False).head(1)
```

```python
occupation = df.loc[(df['salary'] == '>50K') & (df['native-country'] == ' India'), 'occupation'].value_counts()
occupation.sort_values(ascending = False).head(1)
```
